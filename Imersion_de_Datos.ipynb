```python
inmueble['Valor_m2_Millon'] = inmueble['Precio_Millon']/inmueble['Area']
inmueble.head(3)
```

```python
inmueble.groupby('Barrio').mean()
```

```python
datos_barrio = inmueble.groupby('Barrio').sum()
datos_barrio
```

```python
datos_barrio['Valor_m2_Barrio'] = datos_barrio['Precio_Millon']/datos_barrio['Area']
datos_barrio
```

```python
m2_barrio = dict(datos_barrio['Valor_m2_Barrio'])
```

```python
inmueble['Valor_m2_Barrio'] = inmueble['Barrio']
inmueble['Valor_m2_Barrio'] = inmueble['Valor_m2_Barrio'].map(m2_barrio)
inmueble.head(5)
```

```python
top_barrios = inmueble['Barrio'].value_counts()[:10].index
```

```python
datos_barrio.reset_index(inplace=True)
datos_barrio
```

```python
datos_barrio.query('Barrio in @top_barrios')
```

```python
plt.figure(figsize=(16,8))
ax = sns.barplot(x='Barrio',y='Valor_m2_Barrio',data = datos_barrio.query('Barrio in @top_barrios'))
ax.tick_params(axis='x', rotation=45)
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='Barrio',y='Valor_m2_Millon',data = inmueble.query('Barrio in @top_barrios'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='Barrio',y='Valor_m2_Millon',data = inmueble.query('Barrio in @top_barrios & Valor_m2_Millon < 15'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='Barrio',y='Area',data = inmueble.query('Barrio in @top_barrios & Area < 500'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='Barrio',y='Precio_Millon',data = inmueble.query('Barrio in @top_barrios & Precio_Millon < 2000'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```
